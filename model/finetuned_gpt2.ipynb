```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
from summarizer import Summarizer

#load model from drive
#from google.colab import drive
#drive.mount('/content/drive')

#load model from huggingface
from huggingface_hub import notebook_login
notebook_login()
```

```python
# load the pre-trained model+tokenizer from gdrive
#output_dir = '/content/drive/My Drive/opencampus/NFT_text_generation/model'
#model = GPT2LMHeadModel.from_pretrained(output_dir)
#tokenizer = GPT2Tokenizer.from_pretrained(output_dir)

#load model from huggingface
model = GPT2LMHeadModel.from_pretrained("DemocracyStudio/generate_nft_content")
tokenizer = GPT2Tokenizer.from_pretrained("DemocracyStudio/generate_nft_content")

# load the summarizer from library
summarize=Summarizer()

# Tell pytorch to run this model on the GPU.
#device = torch.device("cuda")
#model.cuda()
#model.to(device)
```

```python
prompt = "<|startoftext|>"

generated = torch.tensor(tokenizer.encode(prompt)).unsqueeze(0)
#generated = generated.to(device)

# generate samples from the model
sample_outputs = model.generate(
                                generated, 
                                do_sample=True,   
                                top_k=50, 
                                max_length = 512,
                                top_p=0.95, 
                                num_return_sequences=1
                                )

for i, sample_output in enumerate(sample_outputs):
  # decode model output
  generated_text = tokenizer.decode(sample_output, skip_special_tokens=True)
  print("generated text:", generated_text)
  print("text length: ", len(generated_text))
```

```python
for i, sample_output in enumerate(sample_outputs):
  # generate tweet-length sentence from the model output
  summary = summarize(generated_text, num_sentences=1)
  print("summarizing tweet:", summary)
  print("tweet length: ", len(summary))
```
